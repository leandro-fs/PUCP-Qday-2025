```python
import qiskit as qk 
from qiskit_aer import AerSimulator

# Circuito de 1 qubit en estado |0⟩
qc = qk.QuantumCircuit(1)
stateVectorResult = qk.quantum_info.Statevector(qc)

# Visualización del circuito y estado
display(qc.draw("mpl"))
display(qk.visualization.plot_bloch_multivector(stateVectorResult))

# Agregar medición para obtener histograma
qc_medido = qc.copy()
qc_medido.measure_all()

# Simular con 1000 shots
backend = AerSimulator()
job = backend.run(qc_medido, shots=1000)
result = job.result()
counts = result.get_counts()

# Mostrar histograma
display(qk.visualization.plot_histogram(counts, title="Histograma - Circuito 1 qubit (1000 shots)"))
```

```python
from qiskit.circuit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import *

# Estado de Bell 
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

# Obtener el vector de estado antes de medir
stateVectorResult = Statevector(qc)

# Visualizar esfera de Bloch
display(plot_bloch_multivector(stateVectorResult))

# Agregar mediciones
qc.measure_all()

# Visualizar el circuito completo
display(qc.draw("mpl"))

# Simular con 1000 shots
backend = AerSimulator()
job = backend.run(qc, shots=1000)
result = job.result()
counts = result.get_counts()

# Mostrar histograma
display(plot_histogram(counts, title="Histograma - Estado de Bell (1000 shots)"))
```
